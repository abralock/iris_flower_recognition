```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing  import OneHotEncoder

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
iris_data = load_iris()
```

```python
# Convert to a DataFrame for better visualization
iris_df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
#iris_df['target'] = iris_data.target

# Display the first few rows of the dataset
print(iris_df.head())
```

```python
print("Example data: ")
print(iris_data.data[:5])
```

```python
print("Example labels: ", iris_data.target_names)
print(iris_data.target[:5])
```

```python
x =iris_data.data
y_ = iris_data.target.reshape(-1, 1) # Convert data to a single column
```

```python
# One Hot encode the class labels
encoder = OneHotEncoder(sparse_output=False)
y = encoder.fit_transform(y_)
print(y)
```

```python
# Split the data for training and testing
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.20)
```

```python
# build a model
model = Sequential()
model.add(Dense(10, input_shape=(4,), activation='relu', name="fc1"))
model.add(Dense(10, activation='relu', name="fc2"))
model.add(Dense(3, activation='softmax', name="output"))

print("Neural Network Model Summary: ")
print(model.summary())

```

```python
# Adam optimizer with learning rate of 0.001
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_x, train_y, verbose=2, batch_size=5, epochs=200)

# Test on unseen data
results = model.evaluate(test_x, test_y)
print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
# Adam optimizer with learning rate of 0.001
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(train_x, train_y, verbose=2, batch_size=5, epochs=500)

# Test on unseen data
results = model.evaluate(test_x, test_y)
print('Final test set loss: {:4f}'.format(results[0]))
print('Final test set accuracy: {:4f}'.format(results[1]))
```

```python
for i in range(len(test_x)):
    prediction = np.argmax(model.predict(test_x[i].reshape(1, 4)))
    print("Predicted: {}, Actual: {}".format(prediction, np.argmax(test_y[i])))
```
